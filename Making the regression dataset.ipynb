```python
d1 = pd.read_csv('LICI')
d2 = pd.read_csv('BHARTIARTL')
d3 = pd.read_csv('HDFC')
d4 = pd.read_csv('Reliance')
d5 = pd.read_csv('HINDUNILVR')
d6 = pd.read_csv('SBIN')
d7 = pd.read_csv('ICICIBANK')
d8 = pd.read_csv('TCS')
d9 = pd.read_csv('Infosys')
d10 = pd.read_csv('ITC')
```

```python
d1.dropna(inplace=True)
d2.dropna(inplace=True)
d3.dropna(inplace=True)
d4.dropna(inplace=True)
d5.dropna(inplace=True)
d6.dropna(inplace=True)
d7.dropna(inplace=True)
d8.dropna(inplace=True)
d9.dropna(inplace=True)
d10.dropna(inplace=True)
```

```python
d1['company'] = 'LICI'
d2['company'] ='BHARTIARTL'
d3['company'] = 'HDFC'
d4['company'] = 'Reliance'
d5['company'] = 'HINDUNILVR'
d6['company'] = 'SBIN'
d7['company'] = 'ICICIBANK'
d8['company'] = 'TCS'
d9['company'] = 'Infosys'
d10['company'] = 'ITC'
```

```python
merged_data = pd.concat([d1,d2,d3,d4,d5,d6,d7,d8,d9,d10], axis=0)
```

```python
merged_data = merged_data.sort_values(by='Date').reset_index(drop=True)
```
